### Importando Bibliotecas

```python
import pandas as pd
import matplotlib.pyplot as plt
import os
```

### Pegando número de elementos por subset

```python
arquivo = ["matchups5.csv", "matchups4.csv", "matchups3.csv", "matchups2.csv", "matchups1.csv", "matchupsTodos.csv"]
diretorioArquivo = "../data/entradas_matchups"

tamanho = []
for i in range(len(arquivo)):
    arquivos = os.path.join(diretorioArquivo, arquivo[i])
    df = pd.read_csv(arquivos)
    a = df["Personagem1"]
    tamanho.append(len(a) - 1)

for i in range(len(arquivo)):
    print(tamanho[i])
```

### Lendo arquivo de erros

```python
arquivo = "erros.csv"
diretorioArquivo = "../data/erros"
arquivo = os.path.join(diretorioArquivo, arquivo)
```

```python
df = pd.read_csv(arquivo)
df
```

```python
subsets = df['subset']
erroRelativo = df['erroRelativo']
erroAbsoluto = df['erroAbsoluto']
erroAbsolutoMedio = df['erroAbsolutoMedio']
```

### Montando gráficos

```python
plt.scatter(subsets, erroAbsolutoMedio)
plt.plot(subsets, erroAbsolutoMedio, marker='o', linestyle = '--', color = 'g') 
plt.xlabel('Subsets')         
plt.ylabel('Erro Médio')           
plt.title('Gráfico de Subset X Erro Médio')  
plt.grid(True)                 
plt.show()                   
```

```python
plt.scatter(subsets, erroAbsoluto)
plt.plot(subsets, erroAbsoluto, marker='o', linestyle = '--', color = 'r') 
plt.xlabel('Subsets')            
plt.ylabel('Erro Absoluto')           
plt.title('Gráfico de Subset X Erro Absoluto')  
plt.grid(True)                 
plt.show()               
```

```python
plt.scatter(subsets, erroRelativo)
plt.plot(subsets, erroRelativo, marker='o', linestyle = '--', color = 'b') 
plt.xlabel('Subsets')            
plt.ylabel('Erro Absoluto')           
plt.title('Gráfico de Subset X Erro Relativo')  
plt.grid(True)                 
plt.show()     
```

```python
plt.scatter(subsets, tamanho)
plt.plot(subsets, tamanho, marker='o', linestyle = '--', color = 'y') 
plt.xlabel('Subsets')            
plt.ylabel('Erro Absoluto')           
plt.title('Gráfico de Subset X Quantidade de Itens')  
plt.grid(True)                 
plt.show()     
```
